```python
import pandas as pd

df = pd.read_csv('laptops.csv', sep=';')
df.head()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
df['Cpu_Company'].value_counts()
```

```python
def convert_to_gb(x):
    mem = x.split(' ')[0]
    if mem.endswith('GB'):
        res = mem.replace('GB', '')
    elif mem.endswith('TB'):
        res = float(mem.replace('TB','')) * 1024
    return int(res)
df['Menory_amount'] = df['Memory'].apply(convert_to_gb)
df.head()
```

```python
df['Menory_amount'].value_counts()
```
